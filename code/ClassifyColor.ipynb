```python
import csv
import random

with open('color.txt') as csvfile:
    d = {"red":0,"green":1,"blue":2,"yellow":3}
    data = []
    labels = []
    reader = csv.reader(csvfile, delimiter=',', quotechar=None)
    for row in reader:
        if len(row) == 5:
            data.append(row[0:4])
            labels.append(row[4])
    
    # randomly re-order the data
    ind = range(len(data))
    random.shuffle(ind)
    train = zeros((len(data),4))
    target = empty((len(data),),dtype='a10')
    for to_i,from_i in enumerate(ind):
        train[to_i,:] = asarray(data[from_i])
        target[to_i] = labels[from_i]
```

```python
# map the text labels to integers
import sklearn.preprocessing as prep
le = prep.LabelEncoder()
le.fit(target)
target = le.transform(target)
```
