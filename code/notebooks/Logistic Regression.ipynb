```python
#plot contours of P(Y|x1,x2) for some different models

#1 P(Y|x1,x2) = sigmoid(a+bx2+cx2)

def sigm(t):
    return 1.0/(1.0+exp(-t))

a = 1.0
b = 1.0
c = 1.0
x1 = linspace(-2,2,100)
x2 = linspace(-2,2,100)
X1,X2 = meshgrid(x1,x2)
z = sigm(a+b*X1+c*X2)
fig,ax = subplots(1,1)
ax.spines['left'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position('zero')
ax.spines['top'].set_color('none')
cn = ax.contour(x1,x2,z)
ax.clabel(cn)
show()
```

```python
# Assume that if y = 1, x ~ N1 and if y = 0, x~N2
from scipy.stats import multivariate_normal
pos = empty(X1.shape+(2,))
pos[:,:,0] = X1
pos[:,:,1]= X2
rv = multivariate_normal([1, 1], [[.1, 0], [0, .1]])
rv2 = multivariate_normal([-1, 0], [[2, 0], [0, 2]])
a = rv.pdf(pos)/rv2.pdf(pos)

# generate some data x1,x2,y
n_samples = 1000
data = empty((n_samples,2))
data[0:n_samples/2,:] = rv.rvs(n_samples/2)
data[n_samples/2:,:] = rv2.rvs(n_samples/2)
target = zeros((n_samples,))
target[0:n_samples/2] = 1

# run logistic regression from sklearn
from sklearn.linear_model import LogisticRegression
target[n_samples/2:] = 0
model = LogisticRegression()
model = model.fit(data, target)

fig,ax = subplots(1,4,figsize = (20,5))
[ax[i].spines['left'].set_position('zero') for i in range(0,4)]
[ax[i].spines['right'].set_color('none') for i in range(0,4)]
[ax[i].spines['bottom'].set_position('zero') for i in range(0,4)]
[ax[i].spines['top'].set_color('none') for i in range(0,4)]
cn = ax[0].contourf(x1,x2,rv.pdf(pos))
cn2 = ax[1].contourf(x1,x2,rv2.pdf(pos))
cn3 = ax[2].contourf(x1,x2,a/(1.0+a))

a = model.intercept_[0]
b = model.coef_[0,0]
c = model.coef_[0,1]

colors = ["blue" if x < 1 else "red" for x in target]
x = data[:,0]
y = data[:,1]
ax[3].plot(x1, -(a+b*x1)/c)
ax[3].scatter(x,y,s=30,c=colors,alpha=.5)
show()

# compare error rates of correct classifier with logistic regression

# What does SVM do

# What about if we added higher order features? (then we may need regularization)

# What about the classic ring example?

# Naive Bayes?
```
